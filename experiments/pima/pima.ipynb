```python
import os
data_files = [(f'pima-5-{i}tra', f'pima-5-{i}tst') for i in range(1, 6)]
data_files
```

```python
os.path.join('data', 'pima', '.data')
```

```python
df_train.to_numpy()[:,:8].shape
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

for tr, tst in data_files:
    df_train = pd.read_csv(os.path.join('./data/raw', tr+'.dat'), header=None)
    df_test = pd.read_csv(os.path.join('./data/raw', tst+'.dat'), header=None)
    df_train[8] = df_train[8].apply(str.strip).map({'positive':1, 'negative':0})
    df_test[8] = df_test[8].apply(str.strip).map({'positive':1, 'negative':0})
    df_train[[*range(8)]] = scaler.fit_transform(df_train[[*range(8)]])
    df_test[[*range(8)]] = scaler.transform(df_test[[*range(8)]])
    arr_tr = df_train.to_numpy()
    print(arr_tr.shape)
    arr_tst = df_test.to_numpy()
    print(arr_tst.shape)
    np.savez(os.path.join('data/processed', tr), X=arr_tr[:,:8], y=arr_tr[:,8])
    np.savez(os.path.join('data/processed', tst), X=arr_tst[:, :8], y=arr_tst[:, 8])
```
