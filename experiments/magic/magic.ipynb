```python
balanced = pd.read_csv('./data/full/magic_balanced.csv', header=None)
balanced
```

```python
X = balanced.drop(columns=[10])
y = balanced[10]
scaler = StandardScaler()
X = scaler.fit_transform(X)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, stratify=y)
np.savez('./data/full/magic_train.npz', X=X_train, y=y_train)
np.savez('./data/full/magic_test.npz', X=X_test, y=y_test)
```

# Save Imbalanced Data for different ratios

```python
from datasets import TableDataset
from data_utils import DatasetImbalanced
```

```python
train_file = './data/full/magic_train.npz'
test_file = './data/full/magic_test.npz'
train_data = TableDataset.from_npz(train_file, train=True)
test_data = TableDataset.from_npz(test_file, train=False)
```

```python
# ir 20
ir =100
train_imb = DatasetImbalanced(train_data, imbalance_ratio=1/ir, create_complement_dataset=True)
test_imb = DatasetImbalanced(test_data, imbalance_ratio=1/ir)
```

```python
train_imb.save_npz(f'./data/ir{ir}/magic_train_ir_{ir}.npz')
test_imb.save_npz(f'./data/ir{ir}/magic_test_ir_{ir}.npz')
```

```python
np.savez(f'./data/ir{ir}/magic_train_complement.npz', 
         X=train_imb.complement_dataset.data.numpy(),
         y=train_imb.complement_dataset.target.numpy()
        )
```
